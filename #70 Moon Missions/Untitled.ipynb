```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url="https://en.wikipedia.org/wiki/List_of_missions_to_the_Moon?utm_source=substack&utm_medium=email"
tables = pd.read_html(url, attrs={'class':'wikitable'})
type(tables)
```

```python
print(len(tables))
```

```python
df = pd.DataFrame(tables[1])
df.head(10)
```

```python
df.shape
```

```python
# Remove the last entry of all the unique missions 
unique_mission = df["Mission"].unique()
last_indices = [df[df["Mission"] == mission].index[-1] for mission in unique_mission]
df = df.drop(last_indices)
df.head(10)
```

```python
df["Mission"] = df["Mission"].astype("str")
df["Mission"] = df["Mission"] + " " + df["Mission.1"]
df.head(10)
# df.set_index(["Mission", "Mission.1"], inplace=True)
# df.head(10)
```

```python
df.set_index(["Mission"], inplace=True)
df.head(10)
```

```python
df = df.drop("Mission.1", axis=1)
df.head(10)
```

```python
# Remove all the brackets from some of the date entries
df['Launch date'] = df['Launch date'].str.replace(r'\[.*?\]', '', regex=True)
df.head(20)
```

```python
# Convert launch date to datetime
df["Launch date"] = pd.to_datetime(df["Launch date"])
df.dtypes
```

```python
df[df["Spacecraft"] == "cubesat or similar"]
```

```python
df["Year"] = df["Launch date"].dt.year
df.head()
```

```python
df["Period"] = (df["Year"] // 5) * 5
missions_per_period = df["Period"].value_counts().sort_index()
missions_df = missions_per_period.reset_index()
missions_df.columns = ["Period", "Number of missions"]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=missions_df, x="Period", y="Number of missions", palette='viridis')
plt.title('Number of Space Missions Over 5-Year Periods')
plt.xlabel('5-Year Period')
plt.ylabel('Number of Missions')
plt.show()
```
